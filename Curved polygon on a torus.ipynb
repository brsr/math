```python
#images for the blog post at https://brsr.github.io/2021/05/29/curved-triangle.html
#adapted from https://scipython.com/book/chapter-7-matplotlib/examples/a-torus/
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
%matplotlib notebook
```

```python
c = 1.5 #major radius
a = 1 #minor (tube) radius
n = 100 # number of points
thetalim = -np.pi #"latitude", limited to the lower half for the cross-section
philim = 2*np.pi #"longitude"
```

```python
theta = np.linspace(0, thetalim, n)
phi = np.linspace(0, philim, n)
theta, phi = np.meshgrid(theta, phi)
x = (c + a*np.cos(theta)) * np.cos(phi)
y = (c + a*np.cos(theta)) * np.sin(phi)
z = a * np.sin(theta)
lims = [-a-c, a+c]
popout = (phi < 1*np.pi/2)*1.
```

```python
cmap = matplotlib.colors.ListedColormap(['blue', 'orange'])
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(x, y, z, facecolors=cmap(popout), rstride=5, cstride=5, edgecolors='k', alpha=0.5)
ax.set_xlim3d(lims)
ax.set_ylim3d(lims)
ax.set_zlim3d(lims)
ax.axis('off')
ax.view_init(30, 210)
plt.show()
```
